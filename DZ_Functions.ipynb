# ДЗ Функции

```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

```python
command = ''

while command != 'q':
    command = input('Введите команду: ')

    if command == 'p':
        
        name = input('Введите номер документа: ')
        def search_p(name):
            found = False
            for line in documents:
                if line['number'] == name:
                    found = True
                    print('Владелец документа',line['name'])
            if not found:
                print('Документ не найден в базе')
        search_p(name)
        
    elif command == 's':
        
        name = input('Введите номер документа: ')
        def search_s(name):
            found = False
            for doc in documents:
                if doc['number'] == name:
                    for key in directories.keys():
                        for value in directories[key]:
                            if name == value:
                                found = True
                                print('Документ хранится на полке: ',key)
            if not found:
                print('Документ не найден в базе')
        search_s(name)
        
    elif command == 'l':
        def search_l():
            for _dict in documents:
                for elem in _dict.copy():
                    if elem == 'number':
                        val = _dict[elem]
                        for _list in directories:
                            dir_val = directories[_list]
                            if val in dir_val:
                                _dict['directory'] = _list
            for doc in documents:
                print('№: ',doc['number'], 'тип: ', doc['type'], 'владелец: ', doc['name'], 'полка хранения: ', doc['directory'])
        search_l()
        
    elif command == 'as':
        name = input('Введите номер новой полки: ')
        def search_as(name):
            if name in directories.keys():
                print('Такая полка уже существует. Текущий перечень полок: ')
                for key in directories.keys():
                    print(key)
            else:
                directories[name] = []
                print('Полка добавлена. Текущий перечень полок: ')
                for key in directories.keys():
                    print(key)
        search_as(name)
                
    elif command == 'ad':
        
        name = input('Введите номер полки: ')
        def search_ad(name):
            if name in directories.keys():
                    if len(directories[name]) == 0:
                        del directories[name]
                        print('Полка удалена. Текущий перечень полок:')
                        for key in directories.keys():
                            print(key)
                    else:
                        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ')
                        for key in directories.keys():
                            print(key)
        search_ad(name)
    else:
        print('Game over')
        break


            

```
